```python
#test data
test_input=np.array(data.x[500:700]).reshape([199,1])

#test_output
test_output=np.array(data.y[500:700]).reshape([199,1])
test_output
```

```python
#linear Regression
#step 1-forward propogration

def forward_propogration(train_input,parameters):
    m=parameters['m']
    c=parameters['c']
    
    predictions=np.multiply(m,train_input)+c
    
    
    return predictions
    
    
```

```python
#step-2 Cost Functions
#summession(1/2n*(yi-f(x))^2

def cost_function(predictions,train_output):
    cost=np.mean((train_output-predictions)**2)*0.5
    
    return cost
```

```python
#step-3 gradient descent for backpropagatid
#df=summm(f(x)-y)
#dm=df*x
#dc=df*1
```

```python
def backward_propagation(train_input,train_output,predictions):
    derivaties=dict()
    df=(predictions-train_output)
    dm=np.mean(np.multiply(df,train_input))
    dc=np.mean(df)
    derivaties['dm']=dm
    derivaties['dc']=dc
    
    return derivaties
```

```python
#updated parameters
#m=m-(learning_rate*dm)
#c=c-(learning_rate*dc)
```

```python
def update_parameters(parameters,learning_rate,derivaties):
    parameters['m']=parameters['m']-learning_rate*derivaties['dm'] 
    parameters['c']=parameters['c']-learning_rate*derivaties['dc'] 
    return parameters
    
```

```python
#model training

def train(train_input,train_output,learning_rate,iters):
    parameters=dict()
    parameters['m']=np.random.uniform(0,1)
    parameters['c']=np.random.uniform(0,1)
    
    plt.figure()
    
    loss=list()
    
    for i in range(iters):
        predictions=forward_propogration(train_input,parameters)
        cost=cost_function(predictions,train_output)
        loss.append(cost)
        
        print("Iteration {} lostt {}".format(i,cost))
        
        #plot
        fig,ax=plt.subplots()
        ax.plot(train_input,train_output,"+",label="original")
        ax.plot(train_input,predictions,"*",label="training")
        
        
        legend=ax.legend()
        
        plt.plot(train_input,train_output,"+",label="original")
        plt.plot(train_input,predictions,"*",label="training")
        
        plt.show()
        
        derivaties=backward_propagation(train_input,train_output,predictions)
        parameters=update_parameters(parameters,learning_rate,derivaties)
        
    return parameters,loss
    
```

```python
parameters,loss=train(train_input,train_output,0.0001,20)
print("parameters {} lost{}".format(parameters,loss))
```
